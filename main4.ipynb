```python
import os
import pytesseract
from PIL import Image
```

```python
# Source folder (each image in separate folder)
source_folder = 'img'
```

```python
# Destination folder for moving images
destination_folder = 'images/'
```

```python
# Desired text to check
desired_text = '21/07/23'
```

```python
# Coordinates
x_coordinates = [790, 900]
y_coordinates = [990, 1250]
```

```python
# Loop through each image in the source folder
for filename in os.listdir(source_folder):
    if filename.startswith('frame') and (filename.endswith('.jpg') or filename.endswith('.png')):
        image_path = os.path.join(source_folder, filename)

        # Extract text from image using selected coordinates
        try:
            img = Image.open(image_path)
            extracted_texts = []
            for x, y in zip(x_coordinates, y_coordinates):
                cropped_img = img.crop((x, y, x + 100, y + 30))
                extracted_text = pytesseract.image_to_string(cropped_img)
                extracted_texts.append(extracted_text)
        except Exception as e:
            print("Error while extracting text from", filename, ":", str(e))
            extracted_texts = ["", ""]

        if desired_text in extracted_texts:
            print("Text found in", filename, "- Image will remain in the same location.")
        else:
            new_path = os.path.join(destination_folder, filename)
            os.rename(image_path, new_path)
            print("Text not found in", filename, "- Image moved to destination folder.")

print("Process completed.")
```
